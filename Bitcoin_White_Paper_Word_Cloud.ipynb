```python
#importando pacotes
import textract
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
import nltk.tokenize as word_tokenize
import nltk.corpus as stopwords
from collections import Counter
import string

nltk.download('stopwords')
```

```python
# convertendo pdf para string
def read_file_textract(filepath):
  text = textract.process(filepath)
  return text.decode("utf-8")
```

```python
WhitePaper = "/content/drive/MyDrive/btc white paper/bitcoin.pdf"
texto = read_file_textract(WhitePaper)

WhitePaper_pt = "/content/drive/MyDrive/btc white paper/bitcoin_pt.pdf"
texto_pt = read_file_textract(WhitePaper_pt)
```

```python
# lista de stopword
stopwords = set(STOPWORDS)

# gerar uma wordcloud
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="black",
                      width=1600, height=900).generate(texto)

# mostrar a imagem final
fig, ax = plt.subplots(figsize=(16,9))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()

plt.imshow(wordcloud);
wordcloud.to_file("btc_whitepaper_wordcloud.png")
```

```python
# lista de stopword
stopwords = set(STOPWORDS)
stopwords.update(["da", "meu", "em", "voc", "de", "ao", "os","pode", "cada", 'para'])
stopwords = nltk.corpus.stopwords.words('portuguese')

# gerar uma wordcloud
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="black",
                      width=1600, height=900).generate(texto_pt)

# mostrar a imagem final
fig, ax = plt.subplots(figsize=(16,9))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()

plt.imshow(wordcloud);
wordcloud.to_file("btc_whitepaper_pt_wordcloud.png")
```
